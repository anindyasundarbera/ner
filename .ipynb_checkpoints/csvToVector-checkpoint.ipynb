```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sentence_transformers import SentenceTransformer
import numpy as np
from datetime import datetime
```

```python
# Load the CSV file
file_path = 'event_data.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Check the loaded data
print(df.head())

# Define categories and colors
categories = ['Verb', 'Date', 'Time', 'Country']
colors = ['red', 'blue', 'green', 'purple']
```

```python
# Pre-trained model for text embeddings
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
embedding_dim = model.get_sentence_embedding_dimension()
```

```python

# Function to convert date into enhanced features
def convert_date_enhanced(date_column):
    date_features = []
    for date_str in date_column:
        try:
            # Convert date to datetime object
            date_obj = datetime.strptime(date_str, '%Y-%m-%d')
            # Extract year, month, day, day of the week, day of the year
            year = date_obj.year
            month = date_obj.month
            day = date_obj.day
            day_of_week = date_obj.weekday()
            day_of_year = date_obj.timetuple().tm_yday
            # Cyclic encoding for month and day
            month_sin = np.sin(2 * np.pi * month / 12)
            month_cos = np.cos(2 * np.pi * month / 12)
            day_sin = np.sin(2 * np.pi * day / 31)
            day_cos = np.cos(2 * np.pi * day / 31)
            # Normalize the features
            normalized_year = (year - 2000) / 50  # Example normalization
            normalized_day_of_week = day_of_week / 6
            normalized_day_of_year = day_of_year / 365
            # Create date features
            date_features.append([
                normalized_year, 
                month_sin, month_cos, 
                day_sin, day_cos,
                normalized_day_of_week, 
                normalized_day_of_year
            ])
        except ValueError:
            # Handle any parsing errors with zeros
            date_features.append([0] * 7)
    return np.array(date_features)
```

```python

# Function to convert time into enhanced features
def convert_time_enhanced(time_column):
    time_features = []
    for time_str in time_column:
        try:
            # Convert time to datetime object
            time_obj = datetime.strptime(time_str, '%H:%M')
            # Extract hour and minute
            hour = time_obj.hour
            minute = time_obj.minute
            # Cyclic encoding for hour and minute
            hour_sin = np.sin(2 * np.pi * hour / 24)
            hour_cos = np.cos(2 * np.pi * hour / 24)
            minute_sin = np.sin(2 * np.pi * minute / 60)
            minute_cos = np.cos(2 * np.pi * minute / 60)
            # Create time features
            time_features.append([
                hour_sin, hour_cos,
                minute_sin, minute_cos
            ])
        except ValueError:
            # Handle any parsing errors with zeros
            time_features.append([0] * 4)
    return np.array(time_features)
```

```python
# Convert date and time columns to enhanced numerical features
date_features = convert_date_enhanced(df['Date'])
time_features = convert_time_enhanced(df['Time'])
```

```python
# Embed text columns
verb_embeddings = model.encode(df['Verb'].tolist())
country_embeddings = model.encode(df['Country'].tolist())
```

```python



# Ensure numerical features match embedding dimensions
# Pad date and time features to the same dimension as the text embeddings
def pad_features(features, target_dim):
    num_samples = features.shape[0]
    padded_features = np.zeros((num_samples, target_dim))
    padded_features[:, :features.shape[1]] = features
    return padded_features

# Pad date and time features
date_features_padded = pad_features(date_features, embedding_dim)
time_features_padded = pad_features(time_features, embedding_dim)

# Combine features for PCA with Country at the end
# Stack all features horizontally
combined_features = np.vstack([
    np.hstack([verb_embeddings, np.zeros((len(verb_embeddings), embedding_dim * 3))]),  # Pad for Date, Time, Country
    np.hstack([np.zeros((len(date_features_padded), embedding_dim)), date_features_padded, np.zeros((len(date_features_padded), embedding_dim * 2))]),  # Pad for Verb, Time, Country
    np.hstack([np.zeros((len(time_features_padded), embedding_dim * 2)), time_features_padded, np.zeros((len(time_features_padded), embedding_dim))]),  # Pad for Verb, Date, Country
    np.hstack([np.zeros((len(country_embeddings), embedding_dim * 3)), country_embeddings])  # No padding needed for Verb, Date, Time
])

# Standardize the combined features
scaler = StandardScaler()
combined_features_scaled = scaler.fit_transform(combined_features)

# Apply PCA to the combined features
pca = PCA(n_components=2)
pca_result = pca.fit_transform(combined_features_scaled)

# Calculate the number of samples in each category
num_verb_samples = len(verb_embeddings)
num_date_samples = len(date_features_padded)
num_time_samples = len(time_features_padded)
num_country_samples = len(country_embeddings)

# Plotting the PCA results
plt.figure(figsize=(14, 10))

# Plot Verb category
plt.scatter(pca_result[:num_verb_samples, 0], pca_result[:num_verb_samples, 1], color='red', label='Verb', alpha=0.6, edgecolors='w', s=100)

# Plot Date category
date_start = num_verb_samples
date_end = date_start + num_date_samples
plt.scatter(pca_result[date_start:date_end, 0], pca_result[date_start:date_end, 1], color='blue', label='Date', alpha=0.6, edgecolors='w', s=100)

# Plot Time category
time_start = date_end
time_end = time_start + num_time_samples
plt.scatter(pca_result[time_start:time_end, 0], pca_result[time_start:time_end, 1], color='green', label='Time', alpha=0.6, edgecolors='w', s=100)

# Plot Country category
country_start = time_end
country_end = country_start + num_country_samples
plt.scatter(pca_result[country_start:country_end, 0], pca_result[country_start:country_end, 1], color='purple', label='Country', alpha=0.6, edgecolors='w', s=100)

# Add labels and legend
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('PCA of Text and Temporal Data by Category')
plt.legend()
plt.grid(True)
plt.show()
def convert_date_enhanced(date_column):
    date_features = []
    for date_str in date_column:
        try:
            # Convert date to datetime object
            date_obj = datetime.strptime(date_str, '%Y-%m-%d')
            # Extract year, month, day, day of the week, day of the year
            year = date_obj.year
            month = date_obj.month
            day = date_obj.day
            day_of_week = date_obj.weekday()
            day_of_year = date_obj.timetuple().tm_yday
            # Cyclic encoding for month and day
            month_sin = np.sin(2 * np.pi * month / 12)
            month_cos = np.cos(2 * np.pi * month / 12)
            day_sin = np.sin(2 * np.pi * day / 31)
            day_cos = np.cos(2 * np.pi * day / 31)
            # Normalize the features
            normalized_year = (year - 2000) / 50  # Example normalization
            normalized_day_of_week = day_of_week / 6
            normalized_day_of_year = day_of_year / 365
            # Create date features
            date_features.append([
                normalized_year, 
                month_sin, month_cos, 
                day_sin, day_cos,
                normalized_day_of_week, 
                normalized_day_of_year
            ])
        except ValueError:
            # Handle any parsing errors with zeros
            date_features.append([0] * 7)
    return np.array(date_features)

# Function to convert time into enhanced features
def convert_time_enhanced(time_column):
    time_features = []
    for time_str in time_column:
        try:
            # Convert time to datetime object
            time_obj = datetime.strptime(time_str, '%H:%M')
            # Extract hour and minute
            hour = time_obj.hour
            minute = time_obj.minute
            # Cyclic encoding for hour and minute
            hour_sin = np.sin(2 * np.pi * hour / 24)
            hour_cos = np.cos(2 * np.pi * hour / 24)
            minute_sin = np.sin(2 * np.pi * minute / 60)
            minute_cos = np.cos(2 * np.pi * minute / 60)
            # Create time features
            time_features.append([
                hour_sin, hour_cos,
                minute_sin, minute_cos
            ])
        except ValueError:
            # Handle any parsing errors with zeros
            time_features.append([0] * 4)
    return np.array(time_features)

# Load the pre-trained Sentence Transformers model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Define a function to get vector details for a word
def get_word_vector(word):
    # Encode the word using the Sentence Transformers model
    word_embedding = model.encode(word)
    return word_embedding

# Example usage:
word = 'apple'
word_vector = get_word_vector(word)

# Print the dimensions of the word vector
print(f"Dimensions of the word vector for '{word}': {len(word_vector)}")

# Optionally, print the vector itself (first few elements)
print(f"Vector for '{word}': {word_vector[:10]}")  # Print only the first 10 elements for illustration
```
